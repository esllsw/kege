```python
f = open('26_1.txt')
n = 990
times = []

for i in range(n):
    st,en = list(map(int, f.readline().split()))
    times.append((st,en))

times.sort(key=lambda x: (x[1], x[0]))
print(times)

k = []
max_end = max(en for st, en in times)
sched = [0] * (max_end + 1)
for st,en in times:
    if sum(sched[st:en+1]) == 0:
        k.append(st)
        for t in range(st,en+1):
            sched[t] = 1

print(sched)
print(len(k)+1)
```

```python
f = open('26_2.txt')
n = int(f.readline())

a = [int(x) for x in f.readlines()]
a.sort(reverse=True)
```

```python
gift = [a[0]]
for i in range(n):
    if gift[-1] - a[i] >= 3:
        gift.append(a[i])
print(len(gift),min(gift))
print(gift)
```
