```python
import turtle as t
t.speed(10)
t.screensize(2000,2000)
k = 15
t.down()

for _ in range(8):
    t.forward(12*k)
    t.right(90)

t.up()
for x in range(0,15):
    for y in range(-13,5):
        t.goto(x * k, y *k)
        t.dot()
t.done()
```

Чтобы найти количество точек внутри треугольника, вычислим уравнения прямых, чтобы определить область, в которую они должны попадать. Уравнения прямых найдём через тангенс угла наклона прямых.

y1=tg(30)∗x
y2=tg(150)∗x+b
x=0

Прямая y2 пересекает ось OX в точке (0,30), следовательно, коэффициент b = 30.
Тогда y2=tg(150)∗x+30

Чтобы определить область, перепишем всё в виде неравенств, которым должны удовлетворять наши точки в систему:
​ 
y>tg(30)∗x
y<tg(150)∗x+b
x>0
​
Теперь с помощью питона найдем все точки, которые удовлетворяют этой системе:

```python
from math import sqrt

count = 0  # введём счётчик для количества подходящих точек
for x in range(30):  # цикл перебора значений х
    for y in range(30):  # цикл перебора значений у
        y1 = (sqrt(3) / 3) * x  # уравнение первой прямой
        y2 = -(sqrt(3) / 3) * x + 30  # уравнение второй прямой
        if y > y1 and y < y2 and x > 0:  # условие того, что точка лежит в нужной нам области
            count += 1  # увеличиваем счетчик

print(count)  # выводим ответ
```
